```python
#dff["Network Type"] = dff["Network Type"].dropna().str.replace('G', '').str.replace(" VOLTE",".5")

#dff["Network Type"]

dum = pd.get_dummies(df["Network Type"],drop_first=True)
dff = pd.concat([dum,df],axis = 1)
dff = dff.drop("Network Type", axis=1)
dff
```

```python
dff["Full HD Recording"] = df["Full HD Recording"].fillna(value  = 0).replace("Yes" , "1").replace("No" , "0")
dff["Full HD Recording"]
```

```python
dff["Secondary Camera"] = df["Secondary Camera"].str.replace("MP Front Camera", "")
dff["Secondary Camera"]

```

```python
dff["Secondary Camera Available"] = df["Secondary Camera Available"].replace("Yes","1").replace("No","0").astype(float)
dff["Secondary Camera Available"]
```

```python
dff['Primary Camera Available'] = dff['Primary Camera Available'].replace({'Yes': 1, 'No': 0}).astype(float)
dff['RAM'] = df['RAM'].str.replace(' GB', '').str.replace('MB', '').astype(float)
dff['Internal Storage'] = df['Internal Storage'].str.replace(' MB', '').str.replace('GB', '').astype(float)
dff["Primary Clock Speed"] = dff["Primary Clock Speed"].str.replace("GHz","").str.replace("MHz","").astype(float)
dff["Processor Core"] = dff["Processor Core"].fillna(value = 0).str.replace("Octa Core","8").str.replace("Hexa Core","6").str.replace("Quad Core","4").str.replace("Single Core","1").astype(float)

dummy_variables = pd.get_dummies(dff['Operating System'], prefix='Operating_System: ')
dff = pd.concat([ dummy_variables,dff], axis=1)
dff = dff.drop("Operating System", axis=1)
dff = dff.replace({True: 1, False: 0})
dff

```

```python
# Split the resolution into separate columns
dff["Resolution"] = dff["Resolution"].str.lower().str.replace("X","x").str.replace("$","").str.replace("Ã","").str.replace("*","x").str.replace("-","x").str.replace("pixels","")
dff["Resolution"] = dff["Resolution"].str.replace("pixel" , "").str.replace(" ", "").str.replace(" ", "")#.str.replace(r'\s*x\s*', 'x')
dff[["Width", "Height"]] = dff["Resolution"].str.split('x', expand=True)
dff = dff.drop("Resolution", axis=1)
dff.loc[ 170:,"Width"]

```

```python
dff["Display Size"] = dff["Display Size"].str.split().str[0]
dff["Touchscreen"] = dff["Touchscreen"].str.lower().str.replace("yes" ,"1").str.replace("no","0")
dff["SIM Type"] = dff["SIM Type"].str.replace("Dual Sim","2").str.replace("Single Sim","1").str.replace("Dual Sim(Nano + eSIM)","1.5") 
dummy_variables = pd.get_dummies(dff["Model Name"], prefix='model name: ')
dff = pd.concat([dff, dummy_variables], axis=1)
dff = dff.drop("Model Name",axis=1)

dff["Price"] = dff["Price"].str.replace("₹", "")
dff = dff.replace({True: 1, False: 0})

#dff.to_csv('output.csv', index=False)
dff["Primary Camera"].unique()
dff.isna().sum()

```
